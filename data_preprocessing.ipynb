Processing Match Data

```python
import pandas as pd

# Replace with your actual file path
file_path = '/content/drive/MyDrive/t20_data/match_summary.csv'

df = pd.read_csv(file_path)
df.head()
```

```python
df.shape
```

Use scorecard as a match id to link with other tables

```python
df.rename({'Scorecard': 'match_id'}, axis = 1, inplace = True)
df.head()
```

Create a match ids dictionary that maps team names to a unique match id. This will be useful later on to link with other tables

```python
match_ids_dict = {}

for index, row in df.iterrows():
    key1 = row['Team 1'] + ' Vs ' + row['Team 2']
    key2 = row['Team 2'] + ' Vs ' + row['Team 1']
    match_ids_dict[key1] = row['match_id']
    match_ids_dict[key2] = row['match_id']
```

```python
df.to_csv('/content/drive/MyDrive/t20_data/match_data.csv', index=False)
```
